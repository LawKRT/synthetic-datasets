```python
# Libraries required for the full pipeline
import requests # fetches data from the raw GitHub URL
```

```python
import pandas as pd # converts JSON to DataFrame and exports to CSV
```

```python
import json # parses and previews raw JSON structure
```

```python
# Raw GitHub URL pointing to the uploaded JSON file
url = "https://raw.githubusercontent.com/LawKRT/synthetic-datasets/refs/heads/main/MOCK_DATA.json"
# Send HTTP GET request to the URL
response = requests.get(url)
```

```python
# Confirm the request succeeded
print(f"Status Code: {response.status_code}")
print(f"Content-Type: {response.headers['Content-Type']}")
```

```python
# Deserialize the raw JSON content into a Python list of dicts
data = response.json()
```

```python
# Preview the first record
print(f"Total records: {len(data)}") # Confirm how many records were returned
print(f"First record: {json.dumps(data[0], indent=2)}")
```

```python
# Load the list of dictionaries directly into a pandas DataFrame
df = pd.DataFrame(data)

# Inspect structure
print(df.shape)        # (rows, columns)
print(df.dtypes)       # column data types
df.head()              # preview first 5 rows
```

```python
# Check for nulls(missing values across all columns)
print(df.isnull().sum())

# Specify the exact format from my Mockaroo data (dd/mm/yyyy)
df['hire_date'] = pd.to_datetime(df['hire_date'], format='%d/%m/%Y')

# Confirm salary is numeric
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# Confirm final data types
df.dtypes
```

```python
# Define the full output path
output_path = "mock_employees.csv"

# Save to CSV — index=False removes the auto-generated row numbers
df.to_csv(output_path, index=False)

print(f"CSV saved: {output_path}")
print(f"Rows written: {len(df)}")
```

```python
import requests
import pandas as pd
import json
```

```python
response = requests.get("https://dummyjson.com/products")
raw = response.json()

# Top-level keys
print(raw.keys())
print(f"Total products available: {raw['total']}")
print(f"Records in this response: {len(raw['products'])}")

# Structure of one product
print(json.dumps(raw['products'][0], indent=2))
```

```python
response = requests.get("https://dummyjson.com/carts")
raw = response.json()

print(raw.keys())
print(f"Total carts available: {raw['total']}")
print(f"Records in this response: {len(raw['carts'])}")

print(json.dumps(raw['carts'][0], indent=2))
```

```python
# Default response returns 30 — use limit=0 to fetch all 194
response = requests.get("https://dummyjson.com/products?limit=0")

print(f"Status Code: {response.status_code}")

products_raw = response.json()['products']
print(f"Total products fetched: {len(products_raw)}")
```

```python
# pd.json_normalize handles nested dicts (dimensions, meta) automatically
df_products = pd.json_normalize(products_raw)

# Drop columns that are lists (images, tags, reviews) — not useful in flat CSV
list_columns = [col for col in df_products.columns 
                if df_products[col].apply(lambda x: isinstance(x, list)).any()]

print(f"List-type columns dropped: {list_columns}")
df_products = df_products.drop(columns=list_columns)

print(f"\nShape: {df_products.shape}")
print(f"\nColumns retained:\n{df_products.columns.tolist()}")
df_products.head()
```

```python
# Fix data types
df_products['price']               = pd.to_numeric(df_products['price'], errors='coerce')
df_products['discountPercentage']  = pd.to_numeric(df_products['discountPercentage'], errors='coerce')
df_products['rating']              = pd.to_numeric(df_products['rating'], errors='coerce')
df_products['stock']               = pd.to_numeric(df_products['stock'], errors='coerce')

# Check nulls
print("Null counts:")
print(df_products.isnull().sum())

# Export
products_path = r"C:\Users\mukil\Desktop\Student Files - Power Query Fundamentals\LUXDEV\Python\dummyjson_products.csv"
df_products.to_csv(products_path, index=False)

print(f"\nProducts CSV saved: {products_path}")
print(f"Rows written: {len(df_products)}")
```

```python
response = requests.get("https://dummyjson.com/carts?limit=0")

print(f"Status Code: {response.status_code}")

carts_raw = response.json()['carts']
print(f"Total carts fetched: {len(carts_raw)}")
```

```python
# Extract cart-level fields only (exclude the nested products list)
cart_summary = []

for cart in carts_raw:
    cart_summary.append({
        "cart_id"          : cart['id'],
        "user_id"          : cart['userId'],
        "total_products"   : cart['totalProducts'],
        "total_quantity"   : cart['totalQuantity'],
        "total_price"      : cart['total'],
        "discounted_total" : cart['discountedTotal']
    })

df_carts = pd.DataFrame(cart_summary)

print(f"Shape: {df_carts.shape}")
df_carts.head()
```

```python
# Explode the nested products list so each item gets its own row
cart_items = []

for cart in carts_raw:
    for item in cart['products']:
        cart_items.append({
            "cart_id"            : cart['id'],
            "user_id"            : cart['userId'],
            "product_id"         : item['id'],
            "product_title"      : item['title'],
            "product_price"      : item['price'],
            "quantity"           : item['quantity'],
            "total_item_price"   : item['total'],
            "discount_percentage": item['discountPercentage'],
            "discounted_price"   : item['discountedTotal']
        })

df_cart_items = pd.DataFrame(cart_items)

print(f"Shape: {df_cart_items.shape}")
df_cart_items.head(10)
```

```python
base_path = r"C:\Users\mukil\Desktop\Student Files - Power Query Fundamentals\LUXDEV\Python"

# Cart summary — one row per cart
carts_path = base_path + r"\dummyjson_carts_summary.csv"
df_carts.to_csv(carts_path, index=False)
print(f"Carts summary saved: {carts_path}")
print(f"Rows: {len(df_carts)}")

# Cart items — one row per product per cart
items_path = base_path + r"\dummyjson_cart_items.csv"
df_cart_items.to_csv(items_path, index=False)
print(f"Cart items saved: {items_path}")
print(f"Rows: {len(df_cart_items)}")
```
